## Generating an episode
> Next, we'll explore the process of generating an episode. An episode refers to the interaction between the agent and the environment, starting from the initial state and ending at the terminal state. During an episode, the agent takes actions in each state, interacting with the environment. The episode concludes when the agent reaches the terminal state. In the case of the Frozen Lake environment, the episode ends if the agent reaches either the hole state (H) or the goal state (G), which serve as the terminal states.







> Let's understand how to generate an episode with the random policy. We learned that the random policy selects a random action in each state. So, we will generate an episode by taking random actions in each state. So for each time step in the episode, we take a random action in each state and our episode will end if the agent reaches the terminal state.

** First, let's set the number of time steps:**

```python
num_timesteps = 20
```

```python
for t in range(num_timesteps):
    random_action = env.action_space.sample()
    obs, reward, terminated, truncated , info = env.step(random_action)
```

```python
env = gym.make("FrozenLake-v1", render_mode="human")
state = env.reset()
print('Time Step 0 :')
env.render()
num_timesteps = 20
env.render()
```

```python
num_episodes = 100
num_timesteps = 50
for i in range(num_episodes):
    state = env.reset()
    print('Time Step 0 :')
    env.render()
    print('Starting Episode: ' + str(i))
    for t in range(num_timesteps):
        random_action = env.action_space.sample()
        obs, reward, terminated, truncated , info = env.step(random_action)
        print ('Time Step {} :'.format(t+1))
        env.render()
        if (terminated or truncated):
            if (terminated):
                print ('Terminated episode: ' + str(i) + ' at Time Step {} :'.format(t+1))
            if (truncated):
                print ('truncated episode: ' + str(i) + ' at Time Step {} :'.format(t+1))

            
            break
```
